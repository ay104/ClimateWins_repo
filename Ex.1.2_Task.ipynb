2. Importation of Data

```python
# import data

df_wepredict = pd.read_csv(os.path.join(path,'02 Data Sets','Supervised','Raw Data','Dataset-wepredict-processed.csv'), index_col = False)
```

```python
# inspect dataframe
df_wepredict.head()
```

```python
# change dataframe index to sort index dimensions issue during scaling

df_wepredict.reset_index(inplace=True)
df_wepredict.rename(columns={'index':'id'}, inplace=True)
df_wepredict.head()
```

```python
df_wepredict.info()
```

3. Data Cleaning

```python
# check dataframe for missing values (NaN)

df_wepredict.isnull().sum()
```

No missing values

```python
# check for duplicates

df_wepredict_dupes = df_wepredict[df_wepredict.duplicated()]
```

```python
df_wepredict_dupes
```

No data duplicates

```python
# separate the date column before scaling

date_column = df_wepredict['DATE']  # store date column
numerical_data = df_wepredict.drop('DATE', axis=1) # drop date from data for scaling
```

4. Data Scaling

```python
# define the scaler function (in this data, all columns are numeric and date column is very important to further analysis - no column is dropped)

scaler = StandardScaler()
```

```python
# scales the dataframe and concatenate with the date column

scaled_numerical_data = scaler.fit_transform(numerical_data)                               # scale the numerical data
df_numerical_scaled = pd.DataFrame(scaled_numerical_data, columns = numerical_data.columns) # convert scaled data back to dataframe
df_wepredict_scaled = pd.concat([date_column, df_numerical_scaled], axis=1)                # join back the date column

df_wepredict_scaled.head()
```

```python
# to compare with the unscaled dataframe

df_wepredict.head()
```

5. Export Scaled Data 

```python
# export data and save create any missing directory item

df_wepredict_scaled.to_pickle(os.path.join(path,'02 Data Sets','Supervised','Dataset_wepredict_processed_scaled.pkl'))
```
